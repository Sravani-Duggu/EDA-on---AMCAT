*How have job designations for graduates evolved over the past five years?*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame (make sure to adjust the path accordingly)
# df = pd.read_csv('your_file.csv')  # Uncomment and use your actual file path

# Check and print the column names
print(df.columns)

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Ensure 'DOJ' column is in the correct format and check for correct case
if 'Doj' in df.columns:
    df.rename(columns={'Doj': 'DOJ'}, inplace=True)  # Rename to 'DOJ' if needed

# Convert 'DOJ' to datetime if it's not already
df['DOJ'] = pd.to_datetime(df['DOJ'])

# Extract the year
df['DOJ_year'] = df['DOJ'].dt.year  # Create a new column for the year of joining

# Get unique years available in the dataset
unique_years = df['DOJ_year'].unique()

# Sort the years and take the last 5 years
last_five_years = sorted(unique_years)[-5:]  # Get the last 5 unique years
recent_hires = df[df['DOJ_year'].isin(last_five_years)]

# Count the occurrences of each designation in the last five years
designation_counts = recent_hires['Designation'].value_counts()

# Get the top 10 designations
top_10_designations = designation_counts.nlargest(10).index.tolist()

# Filter the recent hires to include only top 10 designations
filtered_recent_hires = recent_hires[recent_hires['Designation'].isin(top_10_designations)]

# Count the occurrences of each top designation per joining year
designation_trends = filtered_recent_hires.groupby(['DOJ_year', 'Designation']).size().unstack(fill_value=0)

# Plotting the trends
plt.figure(figsize=(12, 6))
for designation in designation_trends.columns:
    plt.plot(designation_trends.index, designation_trends[designation], marker='o', label=designation)

plt.title('Trends of Top 10 Job Designations for Graduates Over the Last Five Years')
plt.xlabel('Year of Joining')
plt.ylabel('Number of Graduates')
plt.xticks(designation_trends.index)  # Set x-axis ticks to be the unique joining years
plt.legend(title='Designation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)  # Optional: Add a grid for better readability
plt.tight_layout()  # Optional: Adjust layout to avoid overlap
plt.show()
```

The graph shows a growing trend towards technical roles among graduates, with Java-related positions being particularly popular. This suggests a positive outlook for career opportunities in the technology industry. Graduates who acquire proficiency in programming, software development, and related skills are likely to have better job prospects and career growth.
